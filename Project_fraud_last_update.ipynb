```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("PaySim-Fraud-Distributed")
    .config("spark.sql.shuffle.partitions", "200")
    .config("spark.driver.memory", "8g")
    .config("spark.executor.memory", "8g")
    .getOrCreate()
)

spark.sparkContext.setLogLevel("ERROR")
```

```python
from pyspark.sql.functions import col

df = spark.read.csv(
    "PS_20174392719_1491204439457_log.csv",
    header=True,
    inferSchema=True
)

df = df.filter(col("type").isin("TRANSFER", "CASH_OUT"))

df = df.select(
    "step", "amount",
    "nameOrig", "oldbalanceOrg", "newbalanceOrig",
    "nameDest", "oldbalanceDest", "newbalanceDest",
    "isFraud"
)

df.write.mode("overwrite").parquet("/content/base_df")
df = spark.read.parquet("/content/base_df")

print("Rows after filter:", df.count())
```

```python
SPLIT_STEP = 400

train_df = df.filter(col("step") <= SPLIT_STEP)
test_df  = df.filter(col("step") >  SPLIT_STEP)

print("Train:", train_df.count())
print("Test :", test_df.count())
```

```python
from pyspark.sql.functions import count, countDistinct, sum as Fsum

out_deg = train_df.groupBy("nameOrig").agg(
    count("*").alias("orig_tx_count"),
    countDistinct("nameDest").alias("orig_unique_dest")
)

in_deg = train_df.groupBy("nameDest").agg(
    count("*").alias("dest_tx_count"),
    countDistinct("nameOrig").alias("dest_unique_orig")
)


out_flow = train_df.groupBy("nameOrig").agg(
    Fsum("amount").alias("orig_sum_amount")
)

in_flow = train_df.groupBy("nameDest").agg(
    Fsum("amount").alias("dest_sum_amount")
)

def apply_degree_flow(df):
    return (
        df.join(out_deg, "nameOrig", "left")
          .join(in_deg, "nameDest", "left")
          .join(out_flow, "nameOrig", "left")
          .join(in_flow, "nameDest", "left")
          .fillna(0)
    )

train_df = apply_degree_flow(train_df)
test_df  = apply_degree_flow(test_df)
```

```python
from pyspark.sql.functions import col, countDistinct
from pyspark.sql.window import Window
from pyspark.sql.functions import count

fraud_art = (
    train_df.filter(col("isFraud") == 1)
            .groupBy("nameOrig")
            .agg(countDistinct("nameDest").alias("fn"))
)

total_art = (
    train_df.groupBy("nameOrig")
            .agg(countDistinct("nameDest").alias("tn"))
)

def apply_fraud_ratio(df):
    return (
        df.join(fraud_art.select("nameOrig", col("fn").alias("__fn")), "nameOrig", "left")
          .join(total_art.select("nameOrig", col("tn").alias("__tn")), "nameOrig", "left")
          .fillna(0)
          .withColumn(
              "orig_fraud_neighbor_ratio",
              col("__fn") / (col("__tn") + 1)
          )
          .drop("__fn", "__tn")
    )

train_df = apply_fraud_ratio(train_df)
test_df  = apply_fraud_ratio(test_df)

def add_balance(df):
    return (
        df.withColumn("orig_diff", col("oldbalanceOrg") - col("newbalanceOrig"))
          .withColumn("dest_diff", col("newbalanceDest") - col("oldbalanceDest"))
          .withColumn("amount_diff_orig", col("amount") - col("orig_diff"))
          .withColumn("amount_diff_dest", col("amount") - col("dest_diff"))
    )

train_df = add_balance(train_df)
test_df  = add_balance(test_df)

WINDOW = 10
w = Window.partitionBy("nameOrig").orderBy("step").rowsBetween(-WINDOW+1, 0)

train_df = train_df.withColumn(
    "tx_burst_score",
    count("step").over(w) / (col("orig_tx_count") + 1)
)

burst_last = (
    train_df.groupBy("nameOrig")
            .agg({"tx_burst_score": "max"})
            .withColumnRenamed("max(tx_burst_score)", "tx_burst_score")
)

test_df = (
    test_df.join(burst_last, "nameOrig", "left")
           .fillna(0)
)
```

```python
from pyspark.sql.functions import (
    col, log1p, sum as Fsum, count as Fcount
)

edges = train_df.select(
    col("nameOrig").alias("src"),
    col("nameDest").alias("dst"),
    col("amount")
)

out_sum = edges.groupBy("src").agg(
    Fsum("amount").alias("out_amt")
)

edges = (
    edges.join(out_sum, "src", "left")
         .withColumn(
             "w",
             col("amount") / (col("out_amt") + 1e-9)
         )
)


node_init = (
    edges.groupBy("src")
         .agg(
             Fcount("*").alias("tx_cnt"),
             Fsum("amount").alias("total_amt")
         )
         .withColumn(
             "suspicious_score",
             log1p(col("tx_cnt")) + log1p(col("total_amt"))
         )
)

Z = node_init.agg(Fsum("suspicious_score")).collect()[0][0]

personal = (
    node_init
        .withColumn("p", col("suspicious_score") / (Z + 1e-9))
        .select(col("src").alias("node"), col("p"))
)

rank = personal.select(
    col("node"),
    col("p").alias("rank")
)

alpha = 0.85
ITER = 6

for _ in range(ITER):
    contrib = (
        edges.join(rank, edges.src == rank.node, "inner")
             .groupBy("dst")
             .agg(Fsum(col("rank") * col("w")).alias("score"))
    )

    rank = (
        contrib.join(personal, contrib.dst == personal.node, "right")
               .fillna(0)
               .select(
                   col("node"),
                   (alpha * col("score") + (1 - alpha) * col("p")).alias("rank")
               )
    )

train_df = (
    train_df.join(
        rank.withColumnRenamed("node", "nameOrig")
            .withColumnRenamed("rank", "orig_pagerank"),
        "nameOrig",
        "left"
    )
    .join(
        rank.withColumnRenamed("node", "nameDest")
            .withColumnRenamed("rank", "dest_pagerank"),
        "nameDest",
        "left"
    )
    .fillna(0)
)

test_df = (
    test_df.join(
        rank.withColumnRenamed("node", "nameOrig")
            .withColumnRenamed("rank", "orig_pagerank"),
        "nameOrig",
        "left"
    )
    .join(
        rank.withColumnRenamed("node", "nameDest")
            .withColumnRenamed("rank", "dest_pagerank"),
        "nameDest",
        "left"
    )
    .fillna(0)
)

```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import (
    col, count, sum as Fsum,
    avg, stddev, lag
)

K = 10
w = (
    Window
    .partitionBy("nameOrig")
    .orderBy("step")
    .rowsBetween(-K + 1, 0)
)
w_lag = (
    Window
    .partitionBy("nameOrig")
    .orderBy("step")
)

def add_temporal_features(df):
    return (
        df
        .withColumn("tx_cnt_10", count("*").over(w))
        .withColumn("tx_amt_sum_10", Fsum("amount").over(w))
        .withColumn("tx_amt_mean_10", avg("amount").over(w))
        .withColumn("tx_amt_std_10", stddev("amount").over(w))
        .withColumn(
            "prev_step",
            lag("step", 1).over(w_lag)
        )
        .withColumn(
            "time_since_last_tx",
            col("step") - col("prev_step")
        )
        .drop("prev_step")
    )

train_df = add_temporal_features(train_df)
test_df  = add_temporal_features(test_df)

train_df = train_df.fillna(0)
test_df  = test_df.fillna(0)
```

```python
from pyspark.sql.functions import col, when, isnan

def apply_strong_contrast_features(df):
    df2 = df

    df2 = df2.withColumn(
        "amount_over_orig_mean",
        col("amount") / (col("orig_sum_amount") / (col("orig_tx_count") + 1))
    )

    df2 = df2.withColumn(
        "amount_over_dest_mean",
        col("amount") / (col("dest_sum_amount") / (col("dest_tx_count") + 1))
    )

    df2 = df2.withColumn(
        "pagerank_ratio",
        col("orig_pagerank") / (col("dest_pagerank") + 1e-6)
    )

    df2 = df2.withColumn(
        "fan_out_ratio",
        col("orig_tx_count") / (col("orig_unique_dest") + 1)
    )

    df2 = df2.withColumn(
        "fan_in_ratio",
        col("dest_tx_count") / (col("dest_unique_orig") + 1)
    )

    df2 = df2.withColumn(
        "amount_zscore_10",
        (col("amount") - col("tx_amt_mean_10")) / (col("tx_amt_std_10") + 1e-6)
    )

    df2 = df2.withColumn(
        "amount_ratio_10",
        col("amount") / (col("tx_amt_mean_10") + 1)
    )

    df2 = df2.withColumn(
        "tx_freq_ratio_10",
        col("tx_cnt_10") / (col("orig_tx_count") + 1)
    )

    for c in [
        "amount_over_orig_mean",
        "amount_over_dest_mean",
        "pagerank_ratio",
        "fan_out_ratio",
        "fan_in_ratio",
        "amount_zscore_10",
        "amount_ratio_10",
        "tx_freq_ratio_10"
    ]:
        df2 = df2.withColumn(
            c,
            when(col(c).isNull() | isnan(col(c)), 0).otherwise(col(c))
        )

    return df2

train_df = apply_strong_contrast_features(train_df)
test_df  = apply_strong_contrast_features(test_df)
```

```python
FEATURES = [
    "step", "amount",
    "orig_diff", "dest_diff",
    "amount_diff_orig", "amount_diff_dest",
    "orig_tx_count", "dest_tx_count",
    "orig_unique_dest", "dest_unique_orig",
    "orig_sum_amount", "dest_sum_amount",
    "orig_pagerank", "dest_pagerank","tx_cnt_10",
    "tx_amt_sum_10","tx_amt_mean_10",
    "tx_amt_std_10","time_since_last_tx",
    "amount_over_orig_mean",
    "amount_over_dest_mean",
    "pagerank_ratio",
    "fan_out_ratio",
    "fan_in_ratio",
    "amount_zscore_10",
    "amount_ratio_10",
    "tx_freq_ratio_10"
]


pdf = (
    train_df
    .select(FEATURES + ["isFraud"])
    .sample(0.3, seed=42)
    .toPandas()
)

X = pdf[FEATURES]
y = pdf["isFraud"]

```

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    precision_recall_curve,
    average_precision_score
)

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
import lightgbm as lgb
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("Train fraud:", y_train.sum())
print("Test fraud :", y_test.sum())
```

```python
neg = (y_train == 0).sum()
pos = (y_train == 1).sum()
scale_pos_weight = neg / pos

print("scale_pos_weight:", scale_pos_weight)
```

```python
models = {
    "XGBoost": XGBClassifier(
        n_estimators=300,
        max_depth=6,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        scale_pos_weight=scale_pos_weight,
        objective="binary:logistic",
        eval_metric="aucpr",
        random_state=42,
        n_jobs=-1
    ),

    "RandomForest": RandomForestClassifier(
        n_estimators=200,
        max_depth=10,
        class_weight="balanced",
        random_state=42,
        n_jobs=-1
    ),

    "LogisticRegression": LogisticRegression(
        class_weight="balanced",
        max_iter=1000,
        solver="liblinear",
        random_state=42
    ),

    "LightGBM": lgb.LGBMClassifier(
        n_estimators=300,
        max_depth=6,
        learning_rate=0.05,
        class_weight="balanced",
        random_state=42
    )
}
```

```python
from sklearn.metrics import (
    precision_recall_curve,
    confusion_matrix,
    precision_score,
    recall_score,
    f1_score,
    average_precision_score
)
import numpy as np
import pandas as pd

results = {}
summary = []

MIN_RECALL = 0.85
MAX_RECALL = 0.97

COST_FN = 100
COST_FP = 1

P = y_test.sum()

for name, model in models.items():
    print(f"\n--- Training {name} ---")

    model.fit(X_train, y_train)
    y_prob = model.predict_proba(X_test)[:, 1]

    precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

    precision = precision[:-1]
    recall = recall[:-1]

    TP = recall * P
    FN = P - TP
    FP = TP * (1 - precision) / (precision + 1e-9)

    valid = (recall >= MIN_RECALL) & (recall <= MAX_RECALL)

    expected_cost = COST_FN * FN + COST_FP * FP
    expected_cost = np.where(valid, expected_cost, np.inf)

    best_idx = np.argmin(expected_cost)

    best_threshold = thresholds[best_idx]
    y_pred = (y_prob >= best_threshold).astype(int)

    cm = confusion_matrix(y_test, y_pred)

    results[name] = {
        "threshold": best_threshold,
        "precision": precision_score(y_test, y_pred),
        "recall": recall_score(y_test, y_pred),
        "f1": f1_score(y_test, y_pred),
        "pr_auc": average_precision_score(y_test, y_prob),
        "confusion_matrix": cm,
        "y_prob": y_prob
    }

    summary.append({
        "Model": name,
        "Threshold": best_threshold,
        "Precision": results[name]["precision"],
        "Recall": results[name]["recall"],
        "F1-score": results[name]["f1"],
        "PR-AUC": results[name]["pr_auc"],
        "TP": cm[1,1],
        "FP": cm[0,1],
        "TN": cm[0,0],
        "FN": cm[1,0],
    })

df_summary = pd.DataFrame(summary).sort_values("Recall", ascending=False)
df_summary
```

```python
summary = []

for name, r in results.items():
    cm = r["confusion_matrix"]
    summary.append({
        "Model": name,
        "Threshold": r["threshold"],
        "Precision": r["precision"],
        "Recall": r["recall"],
        "F1-score": r["f1"],
        "PR-AUC": r["pr_auc"],
        "TP": cm[1,1],
        "FP": cm[0,1],
        "TN": cm[0,0],
        "FN": cm[1,0],
    })

df_summary = pd.DataFrame(summary).sort_values("F1-score", ascending=False)
df_summary
```

```python
def recall_at_k(y_true, y_prob, k):
    n = int(len(y_prob) * k)
    idx = np.argsort(y_prob)[::-1][:n]
    return y_true.iloc[idx].sum() / y_true.sum()


ks = [0.001, 0.002, 0.005, 0.01, 0.02, 0.05]

plt.figure(figsize=(7,5))

for name, r in results.items():
    recalls = [recall_at_k(y_test, r["y_prob"], k) for k in ks]
    plt.plot(
        [k*100 for k in ks],
        recalls,
        marker="o",
        label=name
    )

plt.xlabel("Top K% Transactions Flagged")
plt.ylabel("Recall")
plt.title("Recall @ Top-K%")
plt.legend()
plt.grid()
plt.show()
```

```python
def lift_at_k(y_true, y_prob, k):
    n = int(len(y_prob) * k)
    idx = np.argsort(y_prob)[::-1][:n]
    fraud_rate_k = y_true.iloc[idx].mean()
    base_rate = y_true.mean()
    return fraud_rate_k / base_rate
plt.figure(figsize=(7,5))

for name, r in results.items():
    lifts = [lift_at_k(y_test, r["y_prob"], k) for k in ks]
    plt.plot(
        [k*100 for k in ks],
        lifts,
        marker="o",
        label=name
    )

plt.axhline(1, linestyle="--")  # random baseline
plt.xlabel("Top K% Transactions Flagged")
plt.ylabel("Lift")
plt.title("Lift @ K%")
plt.legend()
plt.grid()
plt.show()
```
